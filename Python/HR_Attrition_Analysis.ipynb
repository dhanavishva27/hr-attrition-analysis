```python
attrition_rate = (df["attrition"].value_counts()["Yes"]/len(df))*100
print(f"Overall Attrition Rate: {attrition_rate:.2f}%")
```

```python
dept_total = (df.groupby("department")["attrition"].count())
## Total No. of People in each Dept
print(dept_total)
```

```python
dept_attrition = df[df["attrition"] == "Yes"].groupby("department")["attrition"].count()
## Total No. of People with Attrition in each Dept
print(dept_attrition)
```

```python
dept_rate = (dept_attrition / dept_total * 100).round(2).sort_values(ascending=False)
```

```python
print(dept_rate)
```

```python
plt.figure(figsize=(8,4))
dept_rate.plot(kind="bar", color=["#E74C3C","#3498DB","#2ECC71"])
plt.title("Attrition Rate by Department")
plt.xlabel("Department")
plt.ylabel("Attrition Rate (%)")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
## sales dept has around 20% Attrition meaning 2/10 employee who joined are leaving the Company. HR Dept has a large number of Employees but it also has a significant amount of employee leaving of 19.05% followed by sales.Then Lastly Research & Development comes with 13.84% which has a total of 63 employees. Research & Development should be focused more because with a minimum Employees but it comes 13.84% Attrition, if not maintained it leads to Major Attrition.
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x="attrition", y="monthlyincome", data=df, palette={"Yes":"#E74C3C","No":"#3498DB"})
plt.title("Monthly Income vs Attrition")
plt.xlabel("Attrition")
plt.ylabel("Monthly Income")
plt.tight_layout()
plt.show()
## Here we can see that the employees having attrition are having the worst Salary. its near 2600 Only but the no category lowest salary is the mdeian of the yes attrition.and the Yes Attrition highest salary is near the no attrition median .
```

```python
labels = ["18-20","21-25","26-35","36-45","46-60"]
df["age_group"] = pd.qcut(df["age"],q=5,labels=labels)
```

```python
df["age_group"]
```

```python
age_attrition = df[df["attrition"]=="Yes"].groupby("age_group")["attrition"].count()
print(age_attrition)
```

```python
age_total = df.groupby("age_group")["attrition"].count()
print(age_total)
```

```python
age_rate = (age_attrition/age_total)* 100
print(age_rate)
```

```python
plt.figure(figsize=(8,4))
age_rate.plot(kind="bar", color=["#E74C3C","#F39C12","#3498DB","#2ECC71"])
plt.title("Attrition Rate by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Attrition Rate (%)")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
## we can clearly see that age group 18 - 20 are leaving more than 25% of the the total company. So we Select very Few Employees with Geat skills in Age 18-20 Group and fcous more on 21-25 Age Group.Since their Attrition Rate is Less compared to 18-20 and 21-25 are Also Fresh Minds and are Innovative.Moving Next to the 26-35 this is the best Age Group our Company should make more Hiring as it is below 10%.
## Moving to next age group 36- 45 this is also one of the best Groups with 10% Attrition.It is the Second Best Age Group. Lastly the age Group of 46-60 which I think we should Hire less as they are having Arrition Rate near 15%.Ofcourse we need people with experience but the age group of 36-45 also gives the experience. The Age Group 18-20 and 21-25 are the ones with more energy and fresh Minds and Innovative.We cant lose them even with High Attrition
```

```python
wlb = df.groupby(["worklifebalance","attrition"]).size().unstack()
```

```python
wlb_clt = wlb.div(wlb.sum(axis=1),axis=0)*100
```

```python
plt.figure(figsize=(8,4))
wlb_clt["Yes"].plot(kind="bar", color="#E74C3C")
plt.title("Attrition Rate by Work Life Balance Score")
plt.xlabel("Work Life Balance (1=Bad, 4=Best)")
plt.ylabel("Attrition Rate (%)")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
## People who gave work life balance 1 are those who dont like the job the Attrition Rate is 30% which is Sensible. Those who gave 2 are somewhat okay with the way the Company flows with about 16%.The next is for 3, They seem Neutral and are okay with Everything like average, they are with the attrition rate of 15% and lasty for the 4 category they should be comfortable with the rating but are somehow wth the second highest Attrtion and serious actions should be taken on that.
```

```python
ot = df.groupby(["overtime","attrition"]).size().unstack()
```

```python
ot_pct = ot.div(ot.sum(axis=0),axis=1)*100
```

```python
plt.figure(figsize=(6,4))
ot_pct["Yes"].plot(kind="bar", color=["#3498DB","#E74C3C"])
plt.title("Attrition Rate by Overtime")
plt.xlabel("Overtime")
plt.ylabel("Attrition Rate (%)")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
## Those with Overtime are above 50% and those who are not are also nearly 45%. I would say that the Attrition Rate by Overtime doesn't affect that much Attrition like they both are almost same.
```
